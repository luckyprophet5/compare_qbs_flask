```python
import pandas as pd
import sqlite3
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_colwidth', None)
```

```python
db = sqlite3.connect("./pbp_data.db")
pbp = pd.read_sql_query("SELECT * FROM play_by_play", db)
```

```python
pbp.head()
```

```python
qb_season_atts = pbp.groupby(['season','passer'], as_index=False).agg({"pass_attempt":"sum"})
qb_season_atts = qb_season_atts.loc[(qb_season_atts.season>=2006) & (qb_season_atts.pass_attempt>=128)].sort_values(['season','pass_attempt'], ascending=[True, False])

def get_passers(year):
    return list(qb_season_atts.loc[qb_season_atts.season==year].passer.unique())

passers_by_year = {year:get_passers(year) for year in range(2006,2020)}
```

```python
import pickle

pickle.dump(passers_by_year, open("passers_by_year.p", "wb"))
```
